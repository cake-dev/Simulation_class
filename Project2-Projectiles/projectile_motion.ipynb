```python
# 1d projectile motion sim in the z direction (up and down) with air resistance
# using the Euler method
import numpy as np
import matplotlib.pyplot as plt

# constants
g = 9.81  # m/s^2
m = 1  # kg
c = 0.1  # drag coefficient
v0 = 10  # m/s
z0 = 0  # m
t0 = 0  # s
tf = 2  # s
dt = 0.01  # s

# time array
t = np.arange(t0, tf, dt)

# initialize arrays
z = np.zeros(len(t))
v = np.zeros(len(t))

# initial conditions
z[0] = z0
v[0] = v0

# Euler method
for i in range(1, len(t)):
    z[i] = z[i-1] + v[i-1]*dt
    v[i] = v[i-1] - g*dt - c/m*v[i-1]*dt

# plot
plt.plot(t, z)
plt.xlabel('Time (s)')
plt.ylabel('Height (m)')
plt.title('Projectile Motion with Air Resistance')
plt.show()
```

```python
# now 2d
# constants
g = 9.81  # m/s^2
m = 1  # kg
c = 0.1  # drag coefficient
v0 = 10  # m/s
x0 = 0  # m
y0 = 0  # m
t0 = 0  # s
tf = 2  # s
dt = 0.01  # s

# time array
t = np.arange(t0, tf, dt)

# initialize arrays
x = np.zeros(len(t))
y = np.zeros(len(t))
vx = np.zeros(len(t))
vy = np.zeros(len(t))

# initial conditions
x[0] = x0
y[0] = y0
vx[0] = v0
vy[0] = v0

# Euler method
for i in range(1, len(t)):
    x[i] = x[i-1] + vx[i-1]*dt
    y[i] = y[i-1] + vy[i-1]*dt
    vx[i] = vx[i-1] - c/m*vx[i-1]*dt
    vy[i] = vy[i-1] - g*dt - c/m*vy[i-1]*dt

# plot, color line as arrows by time
plt.plot(x, y, color='black')
for i in range(0, len(t), 10):
    plt.arrow(x[i], y[i], vx[i]*dt, vy[i]*dt, head_width=0.1, head_length=0.1, color='red')
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.title('Projectile Motion with Air Resistance')
plt.show()
```
